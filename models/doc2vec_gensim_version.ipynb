```python
stopword=[]
result_new=[]
remainderWords=[]
pattern = r"[0-9]"
seg_list = []
flag = 10000

with open('stop_word.txt', 'r', encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        stopword.append(data)

for i in range(len(result)):
    result_new.append(re.sub(pattern,'',str(result[i])))
    seg = jieba.cut(result_new[i],cut_all=True)
    remainderWords = list(filter(lambda a: a not in stopword and a != '\n', seg))
    #lamba 只能有單行的函式
    #filter(function, iterable) 過濾用  前面放判斷函式  後面放可迭代數字
    n=" ".join(remainderWords).split()
    #用空字串join斷完詞的list 再用空格斷開
    seg_list.append(n)
    if i == flag:
        print('{} 篇文章完成'.format(i+1))
        flag+=10000

print(len(seg_list))
```
